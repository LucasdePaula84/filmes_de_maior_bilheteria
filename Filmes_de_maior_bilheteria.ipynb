### Dados retirados do site wikipedia

https://pt.wikipedia.org/wiki/Lista_de_filmes_de_maior_bilheteria

```python
# Importar Biblioteca
import pandas as pd
```

```python
# Extrair os dados da web
tabelas = pd.read_html("https://pt.wikipedia.org/wiki/Lista_de_filmes_de_maior_bilheteria")
```

```python
# Lista com as Tabelas
tabelas
```

```python
# Selecionando a primeira tabela da lista
tabela1 = tabelas[0]
```

```python
# Visualizando a tabela
tabela1
```

```python
# Filtro da tabela com as colunas que eu quero (Diretor x Bilheteria)
tabela_dir_bil = tabela1[["Diretor(a)", "Bilheteria (US$)"]]
tabela_dir_bil.info()
```

```python
# Tratando meus dados e trasformando em valores numericos

tabela_dir_bil["Bilheteria (US$)"] = (
    tabela_dir_bil["Bilheteria (US$)"]
    .str.replace(".", "", regex=False)  # Remove os pontos
    .str.replace(" ", "", regex=False)  # Remove espaços (se existirem)
    .astype("int64")                   # Converte para inteiro
)
tabela_dir_bil.info()
```

```python
# Fazendo o agrupamento por "Diretor(a)"
tabela_dir_bil.groupby("Diretor(a)").sum().sort_values(by="Bilheteria (US$)", ascending=False) 
```

```python
# Fazendo o agrupamento por "Distribuidor"
tabela_dist_bil = tabela1[["Distribuidor", "Bilheteria (US$)"]]

# Tratando meus dados e trasformando em valores numericos

tabela_dist_bil["Bilheteria (US$)"] = (
    tabela_dist_bil["Bilheteria (US$)"]
    .str.replace(".", "", regex=False)  # Remove os pontos
    .str.replace(" ", "", regex=False)  # Remove espaços (se existirem)
    .astype("int64")                   # Converte para inteiro
)


tabela_dist_bil.groupby("Distribuidor").sum().sort_values(by="Bilheteria (US$)", ascending=False) 
```
